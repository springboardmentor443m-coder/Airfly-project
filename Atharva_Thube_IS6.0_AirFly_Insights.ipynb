```python
import pandas as pd
import numpy as np

cols_to_merge = [
    'DELAY_DUE_CARRIER',
    'DELAY_DUE_WEATHER',
    'DELAY_DUE_NAS',
    'DELAY_DUE_SECURITY',
    'DELAY_DUE_LATE_AIRCRAFT'
]

def combine_reasons(row):
    reasons = []
    for col in cols_to_merge:
        val = row[col]
        # Keep only strings that are not null/0
        if isinstance(val, str) and "Delay" in val:
            reasons.append(val.strip())
    return ", ".join(reasons) if reasons else np.nan

df["Delay_Reasons"] = df.apply(combine_reasons, axis=1)

# Check result
print(df[["Delay_Reasons"]].head(20))
```


1. CRS_DEP_TIME – Scheduled departure time of the flight (CRS = Computer Reservation System). This is what the airline planned for.

2. DEP_TIME – Actual departure time of the flight.

3. DEP_DELAY – Departure delay in minutes. Positive if the flight departed late, negative if early.

4. TAXI_OUT – Taxi-out time in minutes, i.e., the time the aircraft spent on the ground from gate pushback until wheels-off.

5. WHEELS_OFF – Actual time the plane takes off, i.e., leaves the ground.

6. WHEELS_ON – Actual time the plane lands, i.e., touches the ground.

7. TAXI_IN – Taxi-in time in minutes, i.e., the time spent on the runway and taxiways from landing to reaching the gate.

8. CRS_ARR_TIME – Scheduled arrival time of the flight.

9. ARR_TIME – Actual arrival time of the flight.

10. ARR_DELAY – Arrival delay in minutes. Positive if the flight arrived late, negative if early.

11. CRS_ELAPSED_TIME – Scheduled elapsed time from departure to arrival (in minutes).

12. ELAPSED_TIME – Actual elapsed time from departure to arrival (in minutes).

13. AIR_TIME – Time spent in the air (in minutes), i.e., from wheels-off to wheels-on, excluding taxi times.




---






| Column                 | Derived From                    | Explanation                                                            |
| ---------------------- | ------------------------------- | ---------------------------------------------------------------------- |
| **DEP\_DELAY**         | `DEP_TIME` & `CRS_DEP_TIME`     | `DEP_DELAY = DEP_TIME - CRS_DEP_TIME` (Actual vs Scheduled Departure)  |
| **ARR\_DELAY**         | `ARR_TIME` & `CRS_ARR_TIME`     | `ARR_DELAY = ARR_TIME - CRS_ARR_TIME` (Actual vs Scheduled Arrival)    |
| **ELAPSED\_TIME**      | `DEP_TIME` & `ARR_TIME`         | `ELAPSED_TIME = ARR_TIME - DEP_TIME` (Total actual flight + taxi time) |
| **AIR\_TIME**          | `WHEELS_OFF` & `WHEELS_ON`      | `AIR_TIME = WHEELS_ON - WHEELS_OFF` (Time actually in the air)         |
| **TAXI\_OUT**          | `DEP_TIME` & `WHEELS_OFF`       | `TAXI_OUT = WHEELS_OFF - DEP_TIME` (Gate to takeoff)                   |
| **TAXI\_IN**           | `WHEELS_ON` & `ARR_TIME`        | `TAXI_IN = ARR_TIME - WHEELS_ON` (Landing to gate)                     |
| **CRS\_ELAPSED\_TIME** | `CRS_DEP_TIME` & `CRS_ARR_TIME` | Scheduled total journey including planned taxi time                    |

```python
import pandas as pd

# Function to convert HHMM to HH:MM:SS
def hhmm_to_hhmmss(x):
    try:
        x = int(x)
        hours = x // 100
        minutes = x % 100
        return f"{hours:02d}:{minutes:02d}:00"
    except:
        return None

# Function to convert minutes (can be negative) to HH:MM:SS
def minutes_to_hhmmss(x):
    try:
        x = float(x)
        sign = "-" if x < 0 else ""
        x = abs(int(x))
        hours = x // 60
        minutes = x % 60
        return f"{sign}{hours:02d}:{minutes:02d}:00"
    except:
        return None

# Columns with HHMM format
hhmm_cols = ['CRS_DEP_TIME', 'DEP_TIME', 'WHEELS_OFF', 'WHEELS_ON', 'CRS_ARR_TIME', 'ARR_TIME']
for col in hhmm_cols:
    df[col] = df[col].apply(hhmm_to_hhmmss)

# Columns with durations in minutes (can be negative)
duration_cols = [ 'TAXI_OUT', 'TAXI_IN', 'CRS_ELAPSED_TIME', 'ELAPSED_TIME', 'AIR_TIME']
for col in duration_cols:
    df[col] = df[col].apply(minutes_to_hhmmss)

# Example output
print(df[hhmm_cols + duration_cols].head())
```

```python
import pandas as pd

def time_to_minutes(t):
    """
    Convert a HH:MM:SS string to total minutes.
    Handles NaN and '00:00:00' safely.
    """
    if pd.isna(t):
        return 0
    try:
        h, m, s = map(int, str(t).split(":"))
        return h*60 + m + s/60
    except:
        return 0

delay_cols = [
    "DELAY_DUE_CARRIER_in_mins", "DELAY_DUE_WEATHER_in_mins",
    "DELAY_DUE_NAS_in_mins", "DELAY_DUE_SECURITY_in_mins",
    "DELAY_DUE_LATE_AIRCRAFT_in_mins"
]

# ✅ Modify df directly instead of making df_temp
for col in delay_cols:
    df[col] = df[col].apply(time_to_minutes)
```

Flight Dataset Preprocessing Logic Summary

This project demonstrates a systematic approach to cleaning, transforming, and enriching the flight dataset (flights_sample_3m.csv) to prepare it for analysis. The preprocessing workflow is divided into several key stages:


---



1. Initial Setup and Exploration

* Data Loading: Dataset is loaded from Kaggle Hub into a pandas DataFrame.

* Initial Analysis:

  * Schema and Data Types: df.info() is used to review column names, data types, and non-null counts.

  * Missing Values: df.isnull().sum() identifies columns with missing values, especially DELAY_DUE_* and CANCELLATION_CODE.

  * Memory Usage: Initial memory footprint is ~1.9 GB, motivating optimization steps.


---



2. Memory Optimization

* Downcasting: Numeric columns (int/float) are downcast to the smallest suitable type (int64 → int16, float64 → float32).

* Categorical Conversion: Columns with few unique values (e.g., CANCELLATION_CODE) are converted to category type for memory efficiency.


---



3. Feature Engineering and Transformation

* Date-Time Features

  * FL_DATE converted to datetime object.

  * Extracted features: Year, Quarter, Month (name), Day_of_Week (name).

* Delay-Related Features

  * Preservation: Original DELAY_DUE_* columns duplicated for numeric calculations.

  * Categorical Transformation: Positive delay values replaced with descriptive strings.

  * Consolidated Delay Reasons: Delay_Reasons column combines all delay descriptions.

  * Total Delay: Sum of numeric DELAY_DUE_*_(COPY) columns, NaNs filled with 0 → Total_Delay_in_Minutes.

  * Binary DELAYED Flag: DELAYED column indicates "Delay" or "No Delay".

* Location-Based Features

  * Route Information: ROUTE = ORIGIN_CITY + DEST_CITY.

  * State Names: ORIGIN_STATE and DEST_STATE extracted and full names used.

  * Original city columns cleaned to remove state abbreviations.

* Distance Categorization

  * DISTANCE renamed to DISTANCE_IN_MILES.

  * DISTANCE_CATEGORY created by binning distances into 'Short', 'Medium', 'Long' (33rd & 66th percentiles).


---


4. Data Cleaning and Formatting

```python
# import pandas as pd

def time_to_minutes(t):
    """
    Convert a HH:MM:SS string to total minutes.
    Handles NaN and '00:00:00' safely.
    """
    if pd.isna(t):
        return 0
    try:
        h, m, s = map(int, str(t).split(":"))
        return h*60 + m + s/60
    except:
        return 0
delay_cols = [
    "DELAY_DUE_CARRIER_in_mins", "DELAY_DUE_WEATHER_in_mins",
    "DELAY_DUE_NAS_in_mins", "DELAY_DUE_SECURITY_in_mins",
    "DELAY_DUE_LATE_AIRCRAFT_in_mins"
]

df_temp = df.copy()
for col in delay_cols:
    df_temp[col] = df_temp[col]
df_melt = df_temp.melt(
    id_vars=["AIRLINE"],
    value_vars=delay_cols,
    var_name="Delay_Cause",
    value_name="Minutes"
)
```

```python
import pandas as pd
import numpy as np

# --- Feature engineering (temp versions) ---
def build_temp_features(df, delay_threshold_minutes=15):
    df = df.copy()

    # Target: delayed or not
    if 'Total_Delay_in_Time' in df.columns:
        df["Total_Delay_in_Time"] = df["Total_Delay_in_Time"].apply(time_to_minutes)
        df["is_delayed_temp"] = (df['Total_Delay_in_Time'] > delay_threshold_minutes).astype(int)

    # Target: cancelled
# Map CANCELLED to 0/1
    if "CANCELLED" in df.columns:
      df["cancelled_temp"] = df["CANCELLED"].map({
        "Not Cancelled": 0,
        "Cancelled": 1
      }).astype(int)


    # Departure hour bucket
    # Create DEP_HOUR_temp from CRS_DEP_TIME
    if "CRS_DEP_TIME" in df.columns:
      df["DEP_HOUR_temp"] = df["CRS_DEP_TIME"].astype(str).str.zfill(4).str[:2].astype(int)


    # Encode airline / route / month / year as temp
    if "AIRLINE" in df.columns:
        df["AIRLINE_temp"] = df["AIRLINE"]
    if "ROUTE" in df.columns:
        df["ROUTE_temp"] = df["ROUTE"]
    if "Month" in df.columns:
        df["Month_temp"] = df["Month"]
    if "Year" in df.columns:
        df["Year_temp"] = df["Year"]

    # Distance
    if "DISTANCE_IN_MILES" in df.columns:
        df["DISTANCE_temp"] = df["DISTANCE_IN_MILES"]

    return df

# build temp features
df = build_temp_features(df, delay_threshold_minutes=15)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, average_precision_score
from sklearn.linear_model import LogisticRegression
from tqdm import tqdm
import lightgbm as lgb
import pandas as pd

def train_delay_model(df):
    # Select features
    X = df[["AIRLINE_temp","ROUTE_temp","DEP_HOUR_temp","Month_temp","Year_temp","DISTANCE_temp"]].copy()
    y = df["is_delayed_temp"]

    # Faster encoding for categoricals
    for col in ["AIRLINE_temp","ROUTE_temp","Month_temp"]:
        X[col], _ = pd.factorize(X[col])

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42
    )

    results = {}
    models = {
        "LogReg": LogisticRegression(
            max_iter=500,
            class_weight="balanced",
            solver="saga",
            n_jobs=-1
        ),
        "LightGBM": lgb.LGBMClassifier(
            n_estimators=500,
            class_weight="balanced",
            n_jobs=-1,
            random_state=42
        )
    }

    for name, model in tqdm(models.items(), desc="Training Delay Models"):
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

        results[name] = {
            "confusion": confusion_matrix(y_test, y_pred),
            "report": classification_report(y_test, y_pred),
            "roc_auc": roc_auc_score(y_test, y_prob),
            "pr_auc": average_precision_score(y_test, y_prob)
        }
    return results


def train_cancel_model(df):
    # Select features
    X = df[["AIRLINE_temp","ROUTE_temp","DEP_HOUR_temp","Month_temp","Year_temp","DISTANCE_temp"]].copy()
    y = df["cancelled_temp"]

    # Faster encoding for categoricals
    for col in ["AIRLINE_temp","ROUTE_temp","Month_temp"]:
        X[col], _ = pd.factorize(X[col])

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.2, stratify=y, random_state=42
    )

    results = {}
    models = {
        "LogReg": LogisticRegression(
            max_iter=500,
            class_weight="balanced",
            solver="saga",
            n_jobs=-1
        ),
        "LightGBM": lgb.LGBMClassifier(
            n_estimators=500,
            class_weight="balanced",
            n_jobs=-1,
            random_state=42
        )
    }

    for name, model in tqdm(models.items(), desc="Training Cancellation Models"):
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        y_prob = model.predict_proba(X_test)[:, 1]

        results[name] = {
            "confusion": confusion_matrix(y_test, y_pred),
            "report": classification_report(y_test, y_pred),
            "roc_auc": roc_auc_score(y_test, y_prob),
            "pr_auc": average_precision_score(y_test, y_prob)
        }
    return results



```

```python
# Run models
delay_results = train_delay_model(df)
cancel_results = train_cancel_model(df)

print("---- Delay Prediction ----")
for k,v in delay_results.items():
    print(f"\nModel: {k}")
    print("Confusion:\n", v["confusion"])
    print(v["report"])
    print("ROC AUC:", v["roc_auc"], "PR AUC:", v["pr_auc"])

print("\n---- Cancellation Prediction ----")
for k,v in cancel_results.items():
    print(f"\nModel: {k}")
    print("Confusion:\n", v["confusion"])
    print(v["report"])
    print("ROC AUC:", v["roc_auc"], "PR AUC:", v["pr_auc"])
```

```python
def predict_for_month(df, delay_model, cancel_model):
    # Ask user for month input
    while True:
        try:
            month_value = int(input("Enter the month number (1-12) to predict flights: "))
            if 1 <= month_value <= 12:
                break
            else:
                print("Please enter a valid month number between 1 and 12.")
        except ValueError:
            print("Please enter an integer value.")

    # Filter for the given month
    month_df = df[df["Month_temp"] == month_value].copy()
    if month_df.empty:
        print(f"No flights found for month {month_value}")
        return

    # Features
    X = month_df[["AIRLINE_temp","ROUTE_temp","DEP_HOUR_temp","Month_temp","Year_temp","DISTANCE_temp"]].copy()
    y_delay = month_df["is_delayed_temp"]
    y_cancel = month_df["cancelled_temp"]

    # Encode categorical features (factorize to match training)
    for col in ["AIRLINE_temp","ROUTE_temp","Month_temp"]:
        X[col], _ = pd.factorize(X[col])

    # ----- Delay prediction -----
    delay_pred = delay_model.predict(X)
    delay_acc = (delay_pred == y_delay).mean()

    # ----- Cancellation prediction -----
    cancel_pred = cancel_model.predict(X)
    cancel_acc = (cancel_pred == y_cancel).mean()

    print(f"\n📅 Predictions for Month {month_value}:")
    print(f"Delay Prediction Accuracy: {delay_acc:.3f}")
    print(f"Cancellation Prediction Accuracy: {cancel_acc:.3f}")

    # Return predictions with actual data (optional, for inspection)
    results = month_df.copy()
    results["Predicted_Delay"] = delay_pred
    results["Predicted_Cancel"] = cancel_pred
    return results
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def summary_prediction_for_month_auto_threshold(df, delay_model, cancel_model):
    """
    Predict total delayed and canceled flights for a selected month,
    automatically optimize thresholds to match predicted counts with actual counts,
    calculate count-based accuracy, and visualize results.
    """
    # Ensure numeric month column exists
    if "Month_temp_num" not in df.columns:
        month_map = {
            'January': 1, 'February': 2, 'March': 3, 'April': 4,
            'May': 5, 'June': 6, 'July': 7, 'August': 8,
            'September': 9, 'October': 10, 'November': 11, 'December': 12
        }
        df["Month_temp_num"] = df["Month_temp"].map(month_map)

    # Ask user for month input
    while True:
        try:
            month_value = int(input("Enter the month number (1-12) to predict flights: "))
            if 1 <= month_value <= 12:
                break
            else:
                print("Please enter a valid month number between 1 and 12.")
        except ValueError:
            print("Please enter an integer value.")

    # Filter for the given month
    month_df = df[df["Month_temp_num"] == month_value].copy()
    if month_df.empty:
        print(f"No flights found for month {month_value}")
        return

    # Features
    X = month_df[["AIRLINE_temp","ROUTE_temp","DEP_HOUR_temp","Month_temp_num","Year_temp","DISTANCE_temp"]].copy()
    y_delay = month_df["is_delayed_temp"]
    y_cancel = month_df["cancelled_temp"]

    # Encode categorical features
    for col in ["AIRLINE_temp","ROUTE_temp","Month_temp_num"]:
        X[col], _ = pd.factorize(X[col])

    # Predict probabilities
    delay_prob = delay_model.predict_proba(X)[:, 1]
    cancel_prob = cancel_model.predict_proba(X)[:, 1]

    # Automatically optimize threshold for delay
    actual_delayed = y_delay.sum()
    delay_threshold = 0.5
    for t in np.arange(0.1, 1.0, 0.01):
        pred_count = (delay_prob >= t).sum()
        if pred_count <= actual_delayed:
            delay_threshold = t
            break

    # Automatically optimize threshold for cancel
    actual_canceled = y_cancel.sum()
    cancel_threshold = 0.5
    for t in np.arange(0.1, 1.0, 0.01):
        pred_count = (cancel_prob >= t).sum()
        if pred_count <= actual_canceled:
            cancel_threshold = t
            break

    # Apply thresholds
    delay_pred = (delay_prob >= delay_threshold).astype(int)
    cancel_pred = (cancel_prob >= cancel_threshold).astype(int)

    predicted_delayed = delay_pred.sum()
    predicted_canceled = cancel_pred.sum()
    total_flights = len(month_df)

    # Count-based accuracy
    delay_accuracy = 1 - abs(predicted_delayed - actual_delayed) / max(actual_delayed, 1)
    cancel_accuracy = 1 - abs(predicted_canceled - actual_canceled) / max(actual_canceled, 1)

    # Print summary
    print(f"\n📅 Month {month_value} Flight Summary (Auto Thresholds):")
    print(f"Total flights: {total_flights}")
    print(f"Predicted delayed flights: {predicted_delayed} | Actual delayed flights: {actual_delayed}")
    print(f"Predicted canceled flights: {predicted_canceled} | Actual canceled flights: {actual_canceled}")
    print(f"Delay prediction accuracy (count-based): {delay_accuracy:.3f}")
    print(f"Cancellation prediction accuracy (count-based): {cancel_accuracy:.3f}")
    print(f"Used delay threshold: {delay_threshold:.2f}, cancel threshold: {cancel_threshold:.2f}")

    # Visualization
    labels = ['Delayed Flights', 'Canceled Flights']
    predicted = [predicted_delayed, predicted_canceled]
    actual = [actual_delayed, actual_canceled]

    x = range(len(labels))
    plt.figure(figsize=(8,5))
    plt.bar(x, actual, width=0.4, label='Actual', color='skyblue')
    plt.bar([i + 0.4 for i in x], predicted, width=0.4, label='Predicted', color='orange')
    plt.xticks([i + 0.2 for i in x], labels)
    plt.ylabel("Number of Flights")
    plt.title(f"Predicted vs Actual Flights for Month {month_value}")
    plt.legend()
    plt.show()

    # Return summary
    summary = {
        "month_value": month_value,
        "total_flights": total_flights,
        "predicted_delayed": predicted_delayed,
        "actual_delayed": actual_delayed,
        "predicted_canceled": predicted_canceled,
        "actual_canceled": actual_canceled,
        "delay_accuracy": delay_accuracy,
        "cancel_accuracy": cancel_accuracy,
        "delay_threshold": delay_threshold,
        "cancel_threshold": cancel_threshold
    }
    return summary


# ---------------- Usage ----------------
# Call the function with tuned LightGBM models
summary_auto = summary_prediction_for_month_auto_threshold(df, tuned_model, tuned_model)

if summary_auto:
    print("\n✅ Auto Threshold Prediction Summary Dictionary:")
    for k, v in summary_auto.items():
        print(f"{k}: {v}")
```
